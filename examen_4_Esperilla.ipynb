 2. Muestra las 5 primeras filas.

 3. Imprime el número total de filas, columnas y el tipo de cada columna.

# **Pregunta 2:**

1. Crea un DataFrame df_filtrado con todas las filas de df_diabetes donde Glucose > 120.

```python
df_filtrado = df_diabetes[df_diabetes['Glucose'] > 120]
df_filtrado.head()
```

 2. Agrega a df_filtrado una columna BMI_Category que clasifique el BMI así:
* "Bajo" si BMI < 18.5
* "Normal" si 18.5 ≤ BMI < 25
* "Sobrepeso" si 25 ≤ BMI < 30
* "Obeso" si BMI ≥ 30

```python
df_filtrado['BMI_Category'] = df_filtrado['BMI'].apply(
    lambda valor: 'Bajo' if valor < 18.5 else
                'Normal' if valor < 25 else
                'Sobrepeso' if valor < 30 else
                'Obeso'
)
```

 3. Muestra las primeras 5 filas de df_filtrado con la nueva columna incluida.



```python
df_filtrado.head()
```

#  **Pregunta 3:**

1. Carga age_groups.csv en un DataFrame df_age_groups.


```python
df_age_groups = pd.read_csv('age_groups.csv')
df_age_groups.head()
```

 2. Haz un merge de df_diabetes con df_age_groups por la columna Age. El resultado se llamará
df_combinado. Debe ser un left merge para mantener todas las filas de df_diabetes.

```python
df_combinado = pd.merge(df_diabetes, df_age_groups, on='Age', how='left')
print("Data Frame Diabetes:",df_diabetes.shape[0])
print("Data Frame Combinado:",df_combinado.shape[0])
```

 3. Muestra las 5 primeras filas de df_combinado para verificar que la columna AgeGroup se haya
agregado correctamente.

```python
df_combinado.head()
```

# **Pregunta 4:**

1. Agrupa df_combinado por AgeGroup.

```python
df_ageGroup = df_combinado.groupby('AgeGroup')
df_ageGroup.head()
```

2. Calcula el promedio de Glucose y el promedio de BMI para cada grupo de edad.


```python
df_glocoseBMI = df_ageGroup.agg({'Glucose': 'mean', 'BMI': 'mean'})
df_glocoseBMI.head()
```

 3. Ordena el resultado por el promedio de Glucose de mayor a menor.


```python
df_glocoseBMI_desc = df_glocoseBMI.sort_values(by='Glucose', ascending = False)
```

 4. Muestra el DataFrame resultante.


```python
df_glocoseBMI_desc.head()
```

#  **Pregunta 6:**

1. Usando df_combinado, crea una tabla pivote que tenga:
* Índice: AgeGroup
